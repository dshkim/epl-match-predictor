```python
# Drop columns that are not needed for prediction
matches_filtered = matches.drop(['comp', 'gf', 'ga', 'xg', 'xga', 'attendance', 'captain', 'match report', 'notes', 'season'], axis=1)
```

```python
matches_filtered.isna().sum()
```

```python
#Drop row with NaN values
matches_cleaned = matches_filtered.dropna()
```

```python
matches_cleaned.isna().sum()
```

```python
matches_cleaned.head()
```

```python
# Mapping dictionary
map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United": "Manchester Utd",
    "Newcastle United": "Newcastle Utd",
    "Tottenham Hotspur": "Tottenham",
    "West Ham United": "West Ham",
    "Wolverhampton Wanderers": "Wolves"
}

# Replace values in both columns using the dictionary to standardize Team names
matches_cleaned['team'] = matches_cleaned['team'].replace(map_values)
matches_cleaned['opponent'] = matches_cleaned['opponent'].replace(map_values)
```

```python
#Convert date to datetime object
matches_cleaned['date'] = pd.to_datetime(matches_cleaned['date'])
```

```python
#Save matches_cleaned DataFrame as a CSV for future use purposes
matches_cleaned.to_csv('epl_matches_cleaned.csv')
```

```python
# Encode categorical features
matches_cleaned = pd.get_dummies(matches_cleaned, columns=['round', 'day', 'venue', 'opponent', 'referee', 'formation', 'opp formation', 'team'])
```

```python
#Keep only hour for time column
matches_cleaned['time'] = matches_cleaned['time'].str.split(':').str[0].astype('int')
```

```python
# Convert result to numerical outcome: 1 for home win, -1 for away win, 0 for draw
matches_cleaned['result'] = matches_cleaned['result'].map({'W': 1, 'L': -1, 'D': 0})
```

```python
matches_cleaned.head()
```

```python
import re

# Define a function to clean column names
def clean_column_names(columns):
    cleaned_columns = []
    for col in columns:
        # Remove unwanted characters using regex
        cleaned_col = re.sub(r'[^\w\s]', '', col)  # Removes all non-alphanumeric characters except whitespace
        cleaned_columns.append(cleaned_col)
    return cleaned_columns

# Apply the cleaning function
matches_cleaned.columns = clean_column_names(matches_cleaned.columns)
```

```python
# Drop duplicated columns based on column names
matches_cleaned = matches_cleaned.loc[:, ~matches_cleaned.columns.duplicated(keep='first')]
```

```python
#Save cleaned DataFrame as a CSV file
matches_cleaned.to_csv('epl_matches_final.csv')
```
